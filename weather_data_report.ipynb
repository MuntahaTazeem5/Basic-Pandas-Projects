***Weather Data Report***

**Scenario: **A city records daily temperature, humidity, and rainfall.

**Tasks & Instructions:**

Use .head() → preview first 5 days.

Use .tail() → preview last 5 days.

Use .describe() → understand variations in temperature & rainfall.

Use .sort_values(by="temperature", ascending=False) → find hottest days.

Use .iloc[100:110] → extract a 10-day range.

```python
import pandas as pd
```

```python
import pandas as pd

data= {
    "Day":pd.date_range(start="2023-01-01", periods=120, freq="D"),
    "temperature":[25 + i%10 for i in range(120)],
    "humidity": [60 + i%20 for i in range(120)],
    "rainfall": [5 if i%7==0 else 0 for i in range(120)]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.sort_values(by="temperature", ascending=False)
```

```python
df.iloc[100:110]
```
