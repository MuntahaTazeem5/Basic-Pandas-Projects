***Hospital Patients Dataset***

**Scenario:** A hospital records patient_id, age, admission_date, and disease.

**Tasks & Instructions:**

Use .info() → inspect null values and datatypes.

Use .describe() → get average age.

Use .sort_values(by="age", ascending=False) → find oldest patients.

Use .head(15) → preview first 15 patients.

Use .iloc[50:55] → check a specific slice.

```python
import pandas as pd
```

```python
data={
    "patient_id":[i for i in range(1,101)],
    "age":[20 + (i % 60) for i in range(100)],
    "admission_date": pd.date_range(start="2023-01-01", periods=100, freq="D"),
    "disease": ["Flu", "Diabetes", "Hypertension", "Covid", "Asthma"] * 20
}
```

```python
df=pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.sort_values(by="age",ascending=False)
```

```python
df.head(15)
```

```python
df.iloc[50:55]
```
