***E-Commerce Product Reviews***

**Scenario:** An e-commerce company stores customer reviews with product names, ratings, and review dates.

 **Tasks & Instructions:**

Use .head() → preview the first 10 reviews.

Use .tail() → check the last 5 reviews (latest reviews).

Use .sort_values(by="rating") → find the lowest-rated products.

Use .iloc[0:5] → extract the first 5 rows manually.

```python
import pandas as pd
```

```python
data={
    "product":["Laptop","Tablet","Mobile","Headphones","Smartwatch","Camera","Speakers"],
    "rating":[4.5, 3.2, 4.0, 2.8, 3.9, 4.7, 3.5],
    "review_date":["2023-01-02","2023-01-05","2023-01-08","2023-01-10","2023-01-12","2023-01-15","2023-01-18"]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sort_values(by="rating")
```

```python
df.iloc[:5]
```
