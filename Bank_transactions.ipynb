***Bank Transactions Dataset***

**Scenario:**A bank has transaction records (customer_id, amount, type, date).

**Tasks & Instructions:**

Use .info() → check datatypes (float, string, datetime).

Use .describe() → summary stats (mean, min, max) of amount.

Use .sort_values(by="amount", ascending=False) → find the highest transactions.

Use .head(20) → preview the top 20 rows.



```python
import pandas as pd
```

```python
data={
    "customer_id":[101, 102, 103, 104, 105],
    "amount": [2500.50, 1200.75, 5000.00, 3100.10, 760.40],
    "type": ["Deposit", "Withdrawal", "Deposit", "Deposit", "Withdrawal"],
    "date": ["2023-02-01","2023-02-02","2023-02-03","2023-02-04","2023-02-05"]

}
```

```python
df=pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.sort_values(by="amount", ascending=False)
```
